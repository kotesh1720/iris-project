### Scatterplot

```python
# list of colors
color = ['red', 'blue', 'green']
species = [ 'Iris-virginica','Iris-versicolor','Iris-setosa']
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['Sepal_length'], x['Sepal_width'], c = color[i], label = species[i])
    plt.title('Scatterplot Sepal')
    plt.xlabel('Sepal-Length')
    plt.ylabel('Sepal-Wibth')
    plt.legend()
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['Petal_length'], x['Petal_width'], c = color[i], label = species[i])
    plt.title('Scatterplot Petal')
    plt.xlabel('Petal-Length')
    plt.ylabel('petal-Wibth')
    plt.legend()
```

# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
X = df2  # input data
Y = df['Species']  # output data 

# split the data from train & split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=.2, random_state=0)
```

```python
x_train.count()
```

```python
x_test.count()
```

### Logistic Regression

```python
ml = LogisticRegression()
ml.fit(x_train,y_train)
```

```python
print('Logistic Regression Accuracy:',ml.score(x_test, y_test))
```

```python
# ml.fit(x_train.values, y_train.values)
# print('Logistic Regression Accuracy:',ml.score(x_test, y_test))
```

### K-nearest neighbors

```python
ml2 = KNeighborsClassifier()
ml2.fit(x_train, y_train)
```

```python
print('K-nearest neighbors Accuracy:',ml2.score(x_test, y_test))
```

```python
# model training 
ml2.fit(x_train.values, y_train.values)
print('K-nearest neighbors Accuracy:',ml2.score(x_test, y_test))
```

### Decision Tree

```python
ml3 = DecisionTreeClassifier()
ml3.fit(x_train, y_train)
```

```python
print('Decision Tree Accuracy:',ml3.score(x_test, y_test))
```

### save the model

```python
# save the model
import  pickle
file_name = 'iris_saved_model'
pickle.dump(ml,open(file_name,'wb'))
```

```python
# save the updated model
try:
    with open(file_name,'wb') as file:
        pickle.dump(ml, file)
        print('Model is saved sucessfully')
except Exception as e:
        print(f'Error:{e}')
```
